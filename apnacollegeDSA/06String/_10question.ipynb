`String.intern()`:

- `intern()` is a method of `java.lang.String`.
- It returns a canonical representation for the string object.
- If the string pool already contains a string equal to this `String` object as determined by `equals(Object)`, then the string from the pool is returned.
- Otherwise, this `String` object is added to the pool and a reference to this object is returned.

Theory:
- Java maintains a special memory area called the string constant pool (part of the method area / metaspace since Java 7+).
- String literals are automatically interned and stored in that pool.
- Calling `intern()` on a `String` ensures that equal strings share the same reference from the pool.
- Example:
    - `String s1 = new String("hello");`
    - `String s2 = s1.intern();`
    - Now `s2 == "hello"` is `true`.
- If a literal `"hello"` already exists in the pool, `intern()` returns that literal reference.
- If not, it adds the current string content to the pool and returns that reference.

Use cases:
- Save memory when many equal strings are created dynamically.
- Enable fast reference equality checks (`==`) for strings that have been interned.
- Useful for symbol tables, parsers, compilers, or deduplication.

Behavior details:
- `intern()` does not change the original `String` object; it returns a pooled reference.
- The returned `String` is from the pool and is immutable.
- `String` is immutable, so interning is safe.
- The pool is global to the JVM and lasts for the lifetime of the JVM.
- Avoid excessive interning in large applications, because the pool can grow and increase memory use.

`StringBuffer`:
- `StringBuffer` has no `intern()` method.
- `StringBuffer` is mutable; its content can change.
- Interning applies only to immutable `String` objects.
- If you need an interned string from a `StringBuffer`, first convert to `String` with `toString()`, then call `intern()`:
    - `String s = stringBuffer.toString().intern();`

Summary:
- `intern()` exists only on `String`.
- It provides a single shared reference for equal string values via the string pool.
- `StringBuffer` cannot be interned directly because it is mutable and has no `intern()` method.

### String.intern() and StringBuffer

- String.intern() returns a canonical (pooled) reference for a String.
- StringBuffer has no intern(); convert to String first with toString() then intern().

```java
// intern() example
String s1 = new String("hello");
String s2 = s1.intern();
String s3 = "hello";

System.out.println(s1 == s2); // false (s1 is a distinct heap object)
System.out.println(s2 == s3); // true  (both reference the pooled literal)
```

```java
// StringBuffer -> intern()
StringBuffer sb = new StringBuffer("world");
String sFromSb = sb.toString().intern();

System.out.println(sFromSb == "world"); // true (pooled literal)
```
