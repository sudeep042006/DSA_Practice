````java
# Why `~5` prints `-6` in Java (not `2`) — and how to get `2`

- Java `int` is a 32-bit two's‑complement value.  
    5 in 32 bits: 00000000 00000000 00000000 00000101  
    `~5` flips every bit: 11111111 11111111 11111111 11111010 — which is the two's‑complement representation of `-6`. Hence `System.out.println(~5);` prints `-6`.

- If you expected `2`, you were likely thinking of flipping only the lower 3 bits of `5` (3‑bit `101` → `010` which is `2`). To flip just the lowest n bits, mask the result:
    - Formula: `(~x) & ((1 << n) - 1)`
    - Example (flip only the lowest 3 bits):
        ```java
        System.out.println(~5);                 // -6 (flip all 32 bits)
        System.out.println((~5) & 0b111);      // 2  (flip only lower 3 bits)
        System.out.println(Integer.toBinaryString((~5) & 0b111)); // "10"
        ```

- About `0b222`: binary literals (prefix `0b`) may contain only `0` or `1`. `0b222` is invalid. Use a binary mask like `0b111` (equals 7) or construct it with `(1<<n)-1`.

Summary: `~5` is `-6` because Java flips all 32 bits. To get the "local" inverted bits you want, apply a mask that keeps only the bit-width you intend.
````
