# Understanding the Prefix Sum

# Prefix Sum Technique

## What is Prefix Sum?

Prefix Sum is a technique where you precompute cumulative sums of elements in an array. For an array `arr`, the prefix sum array `prefix` stores the sum of all elements from index 0 to index i at position i.

**Example:**
- Original array: `[1, 2, 3, 4, 5]`
- Prefix sum array: `[1, 3, 6, 10, 15]`

## When to Use Prefix Sum?

Use Prefix Sum when you need to:

- **Answer range sum queries efficiently** - Calculate the sum of elements between indices i and j in O(1) time after O(n) preprocessing
- **Optimize repeated range calculations** - Avoid recalculating sums for overlapping ranges
- **Handle subarray problems** - Find subarrays with specific sum properties
- **2D matrix operations** - Calculate rectangular region sums quickly
- **Find cumulative statistics** - Track running totals or cumulative frequencies

## Key Benefits

- **Time Complexity:** O(1) per query after O(n) preprocessing
- **Space Complexity:** O(n) for storing prefix sums
- **Use Cases:** Range sum queries, subarray sum problems, 2D matrix sum queries, competitive programming

## Quick Formula

For range sum from index i to j:
```
Sum(i, j) = prefix[j] - prefix[i-1]
```

````java
int nums[] = {1,2,3};
int k = 3;
int prefixSum[] = new int[nums.length];
prefixSum[0] = nums[0];
for(int i = 1; i < nums.length; i++){
    prefixSum[i] = prefixSum[i-1] + nums[i];
}

for(int i = 0; i < prefixSum.length; i++){
    System.out.print(prefixSum[i] + " ");
}

int count = 0;

for(int i = 0; i < prefixSum.length; i++){
    if(prefixSum[i] - k == k){
        count++;
    }
}
System.out.println();
System.out.println(count);
````

````java
int nums[] = {1, 2, 3};
int k = 3;

HashMap<Integer, Integer> map = new HashMap<>();

map.put(0, 1);

int sum = 0;
int count = 0;

for (int i = 0; i < nums.length; i++) {

    sum += nums[i];

    if (map.containsKey(sum - k)) {
        count += map.get(sum - k);
    }

    map.put(sum, map.getOrDefault(sum, 0) + 1);
}

System.out.println(count);
````
