Q1) Zero Sum Sub Array

```java
int arr[] = {6,-1,-3,4,-2,2,4,6,-12,-2};
int total = 0;

for(int i = 0; i < arr.length; i++){
    int sum = 0;
    for( int j = i; j < arr.length; j++){
        sum += arr[j];
        if(sum == 0){
            total++;
        }
    }
}

void printSubarrays(int[] arr) {
    for (int i = 0; i < arr.length; i++) {
        int sum = 0;
        for (int j = i; j < arr.length; j++) {
            sum += arr[j];
            if (sum == 0) {
                System.out.print("[");
                for (int k = i; k <= j; k++) {
                    System.out.print(arr[k]);
                    if (k < j) System.out.print(", ");
                }
                System.out.println("]");
            }
        }
    }
} 

printSubarrays(arr);

System.out.println(total);
```
