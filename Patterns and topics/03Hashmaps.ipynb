```java
int[] nums = {1,1,2,2,3,3,4,4};
HashMap<Integer, Integer> freq = new HashMap<>();

freq.put(0, 1);
count = 1;
for(int i = 1; i < nums.length; i++){
    freq.put(nums[i - 1], count);
    if(freq.containsKey(nums[i])){
        count++;
    }
    freq.put(nums[i - 1], count);

}

System.out.println(freq);
```
